```python
import pandas as pd
```

```python
RUTA = "C:\\Users\\WIN10\\Desktop\\DS_TOTAL\\marketing_campaign.csv"
```

```python
MARK = pd.read_csv(RUTA, sep = "\t")
```

```python
MARK
```

```python
CANTIDAD_M = MARK.groupby("Marital_Status").agg({
    "ID":"count"
})

CANTIDAD_M
```

```python
CANTIDAD_M.plot.bar(y = "ID",
                    width = .7,
                    title = "Estatus de Matrimonios"
                   
                   )
```

```python
MARK
```

```python
EDUC = MARK.groupby("Education").agg({
    "ID":"count"
    
})
EDUC
```

```python
EDUC.plot.bar(y = "ID",
              width = .6,
              title = "Numero de grados por Educacion"
             
             
             )
```
