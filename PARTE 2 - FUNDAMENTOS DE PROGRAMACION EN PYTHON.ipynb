## OPERACIONES DE ASIGNACION

```python
#SUMA DE ASIGNACION 

VARIABLE = 10
VARIABLE -= 5
print(VARIABLE)
```

```python
VARIABLE = 10 
VARIABLE *= 5
print(VARIABLE)
```
